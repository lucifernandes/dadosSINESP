```
import pandas as pd
baseM = pd.read_csv('/content/drive/My Drive/Bases/base_por_municipio1.csv')
baseM = baseM.drop('Unnamed: 0', axis=1)
baseM.columns = ['municipio', 'uf', 'regiao', 'mês', 'vitimas']

baseM = baseM.fillna(0)
baseM['vitimas'] = baseM['vitimas'].astype(int)

baseM = baseM.apply(lambda x: x.replace('NORTE', 'no'))
baseM = baseM.apply(lambda x: x.replace('NORDESTE', 'ne'))
baseM = baseM.apply(lambda x: x.replace('CENTRO-OESTE', 'co'))
baseM = baseM.apply(lambda x: x.replace('SUDESTE', 'se'))
baseM = baseM.apply(lambda x: x.replace('SUL', 'su'))

mes = pd.DataFrame({})
mes = baseM['mês'].astype(str)
mes[mes.str.contains("01-01")]='jan'
mes[mes.str.contains("02-01")]='fev'
mes[mes.str.contains("02-02")]='fev'
mes[mes.str.contains("02-03")]='fev'
mes[mes.str.contains("02-04")]='fev'
mes[mes.str.contains("fev/")]='fev'
mes[mes.str.contains("03-01")]='mar'
mes[mes.str.contains("mar/")]='mar'
mes[mes.str.contains("04-01")]='abr'
mes[mes.str.contains("05-01")]='mai'
mes[mes.str.contains("06-01")]='jun'
mes[mes.str.contains("07-01")]='jul'
mes[mes.str.contains("08-01")]='ago'
mes[mes.str.contains("09-01")]='set'
mes[mes.str.contains("10-01")]='out'
mes[mes.str.contains("11-01")]='nov'
mes[mes.str.contains("12-01")]='dez'
mes[mes.str.contains("dez/")]='dez'
baseM['mes'] = mes




baseM = baseM.apply(lambda x: x.astype(str).str.lower())

def corrigir_nomes(nome):
    nome = nome.replace('á', 'a').replace('â', 'a').replace('ã','a').replace('ó', 'o').replace('ô', 'o').replace('õ','o').replace('é', 'e').replace('ê', 'e').replace(' ','').replace('í', 'i').replace('ú', 'u').replace('ç','c')
    return nome
baseM['municipio'] = baseM['municipio'].apply(corrigir_nomes)

ano = pd.DataFrame({})
ano = baseM['mês'].astype(str)
ano[ano.str.contains("2018")]='2018'
ano[ano.str.contains("19")]='2019'
ano[ano.str.contains("2020")]='2020'
ano[ano.str.contains("/20")]='2020'
ano = ano.astype(int)
baseM['ano'] = ano

baseM = baseM.drop('mês', axis=1)

municipio = pd.DataFrame({})
municipio['regiao'] = baseM['regiao']
municipio['uf'] = baseM['uf']
municipio['ano'] = baseM['ano']
municipio['mes'] = baseM['mes']
municipio['municipio'] = baseM['municipio']
municipio['vitimas_municipio'] = baseM['vitimas']

#municipio.to_csv('b_municipio.csv')
municipio.head(20000)
```
